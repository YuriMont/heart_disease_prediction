### Treinamento de Modelo: Random Forest com GridSearchCV

Agora vamos implementar um modelo de Classificação usando Random Forest e otimizá-lo com `GridSearchCV` para encontrar os melhores hiperparâmetros.


**OBS**: `X_train, X_test, y_train, y_test` são importadas emboras o editor o considere inexistente

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, accuracy_score

# 1. Definir o modelo (KNeighborsClassifier)
knn = KNeighborsClassifier()

# 2. Definir o grid de parâmetros para o GridSearchCV
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11, 13, 15],  # Número de vizinhos
    'weights': ['uniform', 'distance'],      # Função de peso usada na previsão
    'metric': ['euclidean', 'manhattan']     # Métrica de distância
}

# 3. Inicializar GridSearchCV
# cv=5 indica 5-fold cross-validation
# scoring='accuracy' indica que a métrica de avaliação será acurácia
# n_jobs=-1 usa todos os processadores disponíveis
grid_search = GridSearchCV(estimator=knn, param_grid=param_grid,
                           cv=5, scoring='accuracy', n_jobs=-1, verbose=1)

print("Iniciando GridSearchCV para KNN...")
# 4. Ajustar (fit) o GridSearchCV aos dados de treino
grid_search.fit(X_train, y_train)
print("GridSearchCV concluído!")
```

### Resultados do GridSearchCV

Vamos inspecionar os melhores parâmetros e a melhor pontuação encontrados pelo `GridSearchCV`.

```python
# 5. Exibir os melhores parâmetros e a melhor pontuação
print(f"Melhores parâmetros encontrados: {grid_search.best_params_}")
print(f"Melhor pontuação (acurácia) no treino: {grid_search.best_score_:.4f}")

# 6. Obter o melhor modelo
best_knn_model = grid_search.best_estimator_

# Fazer previsões no conjunto de teste com o melhor modelo
y_pred = best_knn_model.predict(X_test)

# 7. Avaliar o desempenho do modelo no conjunto de teste
print("\nRelatório de Classificação no conjunto de teste:")
print(classification_report(y_test, y_pred))

print(f"Acurácia do melhor modelo no conjunto de teste: {accuracy_score(y_test, y_pred):.4f}")
```

### Matriz de Confusão e Métricas de Desempenho

A Matriz de Confusão é uma ferramenta fundamental para avaliar o desempenho de um modelo de classificação. Ela mostra o número de acertos e erros do modelo, categorizando-os em Verdadeiros Positivos, Verdadeiros Negativos, Falsos Positivos e Falsos Negativos.

```python
from sklearn.metrics import (
    accuracy_score, recall_score, precision_score, f1_score,
    confusion_matrix, roc_curve, auc, ConfusionMatrixDisplay
)
import matplotlib.pyplot as plt

# Gerar a matriz de confusão
cm = confusion_matrix(y_test, y_pred, labels=best_knn_model.classes_)

# Exibir a matriz de confusão
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=best_knn_model.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.title('Matriz de Confusão para KNN')
plt.show()

# O relatório de classificação já foi exibido, mas podemos reforçar algumas métricas aqui.
# Por exemplo, calculando a acurácia, precisão, recall e f1-score manualmente se necessário,
# ou apenas relembrando os valores do classification_report.

print("\nRelembrando as métricas do modelo:")
print(f"Acurácia: {accuracy_score(y_test, y_pred):.4f}")

# Podemos também extrair precisão, recall, f1-score para cada classe
# No caso do GridSearchCV, o best_estimator_ já está treinado.
# Vamos apenas usar as previsões que já foram feitas para y_pred

accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("\nDesempenho no conjunto de teste:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Recall (Sensibilidade): {recall:.4f}")
print(f"Precision: {precision:.4f}")
print(f"F1-Score: {f1:.4f}")
```

### Entendendo a Matriz de Confusão:

*   **Verdadeiros Positivos (VP):** Casos onde o modelo previu corretamente a classe positiva (ex: previu doença, e o paciente tem doença).
*   **Verdadeiros Negativos (VN):** Casos onde o modelo previu corretamente a classe negativa (ex: previu ausência de doença, e o paciente não tem doença).
*   **Falsos Positivos (FP):** Casos onde o modelo previu a classe positiva incorretamente (ex: previu doença, mas o paciente não tem doença - Erro Tipo I).
*   **Falsos Negativos (FN):** Casos onde o modelo previu a classe negativa incorretamente (ex: previu ausência de doença, mas o paciente tem doença - Erro Tipo II).

### Métricas Derivadas:

*   **Acurácia:** `(VP + VN) / (VP + VN + FP + FN)` - Proporção de previsões corretas sobre o total de previsões.
*   **Precisão (Precision):** `VP / (VP + FP)` - Das vezes que o modelo previu a classe positiva, quantas ele acertou. Importante quando o custo de um Falso Positivo é alto.
*   **Recall (Sensibilidade):** `VP / (VP + FN)` - Dos casos reais da classe positiva, quantos o modelo conseguiu identificar. Importante quando o custo de um Falso Negativo é alto.
*   **F1-Score:** `2 * (Precisão * Recall) / (Precisão + Recall)` - Média harmônica da precisão e do recall, útil quando há um desequilíbrio entre essas duas métricas.

### Curva ROC (Receiver Operating Characteristic) e AUC (Area Under the Curve)

A Curva ROC é um gráfico que mostra a taxa de verdadeiros positivos (TPR) versus a taxa de falsos positivos (FPR) para diferentes pontos de corte. A AUC é a área sob a Curva ROC, fornecendo uma métrica agregada do desempenho de classificação em todos os possíveis pontos de corte. Uma AUC de 1.0 representa um modelo perfeito, enquanto uma AUC de 0.5 representa um modelo que não é melhor que um classificador aleatório.

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Obter as probabilidades de previsão para a classe positiva (classe 1)
# best_knn_model.predict_proba retorna as probabilidades para todas as classes.
# Precisamos da probabilidade da classe positiva (índice 1).
y_pred_proba = best_knn_model.predict_proba(X_test)[:, 1]

# Calcular a Curva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# Calcular a AUC (Área Sob a Curva)
roc_auc = auc(fpr, tpr)

# Plotar a Curva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'Curva ROC KNN (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Classificador Aleatório')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taxa de Falsos Positivos (FPR)')
plt.ylabel('Taxa de Verdadeiros Positivos (TPR)')
plt.title('Curva ROC para KNN')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

print(f"A Área Sob a Curva (AUC) do modelo KNN é: {roc_auc:.4f}")
```

### Salvando o Modelo Treinado

É uma boa prática salvar o modelo treinado para que ele possa ser reutilizado sem a necessidade de reexecutar todo o processo de treinamento. Usaremos a biblioteca `joblib` para serializar o modelo.

```python
import joblib
import os

# Certifique-se de que project_base_path esteja definido ou obtenha-o do contexto
# Neste caso, vamos assumir que project_base_path está definido a partir de uma célula anterior.

# 1. Definir o caminho para a pasta 'packages'
packages_path = os.path.join(project_base_path, 'packages')

# 2. Criar a pasta 'packages' se ela não existir
if not os.path.exists(packages_path):
    os.makedirs(packages_path)
    print(f"Pasta '{packages_path}' criada.")

# 3. Definir o caminho completo para o arquivo do modelo
model_filename = os.path.join(packages_path, 'knn_model.joblib')

# 4. Salvar o modelo
joblib.dump(best_knn_model, model_filename)
print(f"Modelo salvo com sucesso em: {model_filename}")
```
