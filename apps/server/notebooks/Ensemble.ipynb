### Treinamento de Modelo: Random Forest com GridSearchCV

Agora vamos implementar um modelo de Classificação usando Random Forest e otimizá-lo com `GridSearchCV` para encontrar os melhores hiperparâmetros.


**OBS**: `X_train, X_test, y_train, y_test` são importadas emboras o editor o considere inexistente

```python
import joblib
import os
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import classification_report, accuracy_score

# Define the base path for loading models
# project_base_path should be defined in a previous cell

# Define paths for the pre-trained models
knn_model_path = os.path.join(project_base_path, 'packages', 'knn_model.joblib')
svm_model_path = os.path.join(project_base_path, 'packages', 'svm_model.joblib')
rf_model_path = os.path.join(project_base_path, 'packages', 'random_forest_model.joblib')

# Load the pre-trained models
print(f"Carregando modelo KNN de: {knn_model_path}")
knn_model = joblib.load(knn_model_path)
print(f"Carregando modelo SVM de: {svm_model_path}")
svm_model = joblib.load(svm_model_path)
print(f"Carregando modelo Random Forest de: {rf_model_path}")
random_forest_model = joblib.load(rf_model_path)

print("Modelos carregados com sucesso!")

# Criar o ensemble com VotingClassifier
# Usamos 'soft' voting para que as probabilidades de cada modelo sejam consideradas
ensemble_model = VotingClassifier(
    estimators=[
        ('knn', knn_model),
        ('svm', svm_model),
        ('rf', random_forest_model)
    ],
    voting='soft' # Use 'soft' voting for probability-based predictions
)

# O VotingClassifier não precisa ser "treinado" novamente se os modelos base já estão treinados.
# No entanto, ele precisa ser "fit" aos dados de treino para ter acesso aos nomes das classes (classes_).
# Isso é importante para funções como predict_proba.
# Se os modelos base já foram treinados, o .fit() do VotingClassifier pode ser chamado sem retreinar os modelos base.
# Para garantir que os modelos internos não sejam retreinados, podemos configurá-los como pré-treinados.
# No entanto, a forma mais comum é simplesmente chamar .fit() com os modelos já treinados.
# O `voting='soft'` requer que os estimadores suportem `predict_proba`.

# Ajustar o ensemble (apenas para coletar as classes e outras propriedades do fit)
print("Ajustando o ensemble (VotingClassifier) aos dados de treino...")
ensemble_model.fit(X_train, y_train)
print("Ensemble ajustado!")
```

### Matriz de Confusão e Métricas de Desempenho

A Matriz de Confusão é uma ferramenta fundamental para avaliar o desempenho de um modelo de classificação. Ela mostra o número de acertos e erros do modelo, categorizando-os em Verdadeiros Positivos, Verdadeiros Negativos, Falsos Positivos e Falsos Negativos.

```python
from sklearn.metrics import (
    accuracy_score, recall_score, precision_score, f1_score,
    confusion_matrix, roc_curve, auc, ConfusionMatrixDisplay
)
import matplotlib.pyplot as plt

# Gerar a matriz de confusão para o modelo ensemble
cm_ensemble = confusion_matrix(y_test, y_pred_ensemble, labels=best_ensemble_model.classes_)

# Exibir a matriz de confusão
disp_ensemble = ConfusionMatrixDisplay(confusion_matrix=cm_ensemble, display_labels=best_ensemble_model.classes_)
disp_ensemble.plot(cmap=plt.cm.Blues)
plt.title('Matriz de Confusão para o Modelo Ensemble')
plt.show()

print("\nRelembrando as métricas do modelo Ensemble:")
print(f"Acurácia: {accuracy_score(y_test, y_pred_ensemble):.4f}")

accuracy_ensemble = accuracy_score(y_test, y_pred_ensemble)
recall_ensemble = recall_score(y_test, y_pred_ensemble)
precision_ensemble = precision_score(y_test, y_pred_ensemble)
f1_ensemble = f1_score(y_test, y_pred_ensemble)

print("\nDesempenho do Modelo Ensemble no conjunto de teste:")
print(f"Accuracy: {accuracy_ensemble:.4f}")
print(f"Recall (Sensibilidade): {recall_ensemble:.4f}")
print(f"Precision: {precision_ensemble:.4f}")
print(f"F1-Score: {f1_ensemble:.4f}")
```

### Entendendo a Matriz de Confusão:

*   **Verdadeiros Positivos (VP):** Casos onde o modelo previu corretamente a classe positiva (ex: previu doença, e o paciente tem doença).
*   **Verdadeiros Negativos (VN):** Casos onde o modelo previu corretamente a classe negativa (ex: previu ausência de doença, e o paciente não tem doença).
*   **Falsos Positivos (FP):** Casos onde o modelo previu a classe positiva incorretamente (ex: previu doença, mas o paciente não tem doença - Erro Tipo I).
*   **Falsos Negativos (FN):** Casos onde o modelo previu a classe negativa incorretamente (ex: previu ausência de doença, mas o paciente tem doença - Erro Tipo II).

### Métricas Derivadas:

*   **Acurácia:** `(VP + VN) / (VP + VN + FP + FN)` - Proporção de previsões corretas sobre o total de previsões.
*   **Precisão (Precision):** `VP / (VP + FP)` - Das vezes que o modelo previu a classe positiva, quantas ele acertou. Importante quando o custo de um Falso Positivo é alto.
*   **Recall (Sensibilidade):** `VP / (VP + FN)` - Dos casos reais da classe positiva, quantos o modelo conseguiu identificar. Importante quando o custo de um Falso Negativo é alto.
*   **F1-Score:** `2 * (Precisão * Recall) / (Precisão + Recall)` - Média harmônica da precisão e do recall, útil quando há um desequilíbrio entre essas duas métricas.

### Curva ROC (Receiver Operating Characteristic) e AUC (Area Under the Curve)

A Curva ROC é um gráfico que mostra a taxa de verdadeiros positivos (TPR) versus a taxa de falsos positivos (FPR) para diferentes pontos de corte. A AUC é a área sob a Curva ROC, fornecendo uma métrica agregada do desempenho de classificação em todos os possíveis pontos de corte. Uma AUC de 1.0 representa um modelo perfeito, enquanto uma AUC de 0.5 representa um modelo que não é melhor que um classificador aleatório.

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Obter as probabilidades de previsão para a classe positiva (classe 1) do ensemble
# best_ensemble_model.predict_proba retorna as probabilidades para todas as classes.
# Precisamos da probabilidade da classe positiva (índice 1).
y_pred_proba_ensemble = best_ensemble_model.predict_proba(X_test)[:, 1]

# Calcular a Curva ROC para o ensemble
fpr_ensemble, tpr_ensemble, thresholds_ensemble = roc_curve(y_test, y_pred_proba_ensemble)

# Calcular a AUC (Área Sob a Curva) para o ensemble
roc_auc_ensemble = auc(fpr_ensemble, tpr_ensemble)

# Plotar a Curva ROC do ensemble
plt.figure(figsize=(8, 6))
plt.plot(fpr_ensemble, tpr_ensemble, color='darkorange', lw=2, label=f'Curva ROC Ensemble (AUC = {roc_auc_ensemble:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Classificador Aleatório')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taxa de Falsos Positivos (FPR)')
plt.ylabel('Taxa de Verdadeiros Positivos (TPR)')
plt.title('Curva ROC para o Modelo Ensemble')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

print(f"A Área Sob a Curva (AUC) do modelo Ensemble é: {roc_auc_ensemble:.4f}")
```

### Salvando o Modelo Treinado

É uma boa prática salvar o modelo treinado para que ele possa ser reutilizado sem a necessidade de reexecutar todo o processo de treinamento. Usaremos a biblioteca `joblib` para serializar o modelo.

```python
import joblib
import os

# Certifique-se de que project_base_path esteja definido ou obtenha-o do contexto
# Neste caso, vamos assumir que project_base_path está definido a partir de uma célula anterior.

# 1. Definir o caminho para a pasta 'packages'
packages_path = os.path.join(project_base_path, 'packages')

# 2. Criar a pasta 'packages' se ela não existir
if not os.path.exists(packages_path):
    os.makedirs(packages_path)
    print(f"Pasta '{packages_path}' criada.")

# 3. Definir o caminho completo para o arquivo do modelo ensemble
ensemble_model_filename = os.path.join(packages_path, 'ensemble_voting_model.joblib')

# 4. Salvar o modelo ensemble
joblib.dump(best_ensemble_model, ensemble_model_filename)
print(f"Modelo Ensemble salvo com sucesso em: {ensemble_model_filename}")
```
